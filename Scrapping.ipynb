## Pagination

```python
titles=[]
ranks=[]
ratings=[]
cities=[]
p=1
while p<=201:
    time.sleep(3)
    page = requests.get('https://vuzoteka.ru/%D0%B2%D1%83%D0%B7%D1%8B?page={}'.format(p)).text
    soup = BeautifulSoup(page, 'html.parser')
    appeals=soup.find_all(class_='institute-search-wrapper')
    for i in range(len(appeals)):
        a=appeals[i]
        titles.append(a.find_all(class_='institute-search-title')[0].get_text())
        cities.append(a.find_all(class_='institute-search-value')[1].get_text())
        ranks.append(a.find_all(class_='institute-search-value')[3].get_text().split()[0])
        ratings.append(a.find_all(class_='institute-search-value')[3].get_text().split()[1])
    p+=1
```

```python
data = pd.DataFrame()
data['title']=titles
data['city']=cities
data['rank']=ranks
data['rating']=ratings
data
```

```python
data.to_csv(r'uni_russia_ranks.csv',index=False)
```
